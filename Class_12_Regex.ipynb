```python
#Write a Python program which takes a list of numbers as input and prints a new list after removing even numbers from it.
#Input:- [10,21,22,98,87,45,33,1,2,100]
#Output:- [21,87,45,33,1]

input_lst =  [10,21,22,98,87,45,33,1,2,100]
out_lst = []

for index in range(len(input_lst)):
    if input_lst[index]%2!=0:
        out_lst.append(input_lst[index])

print(out_lst)
```

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[A-Z]      - Matches Characters in brackets
[^A-Z ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```

```python
match 
    - looks for the pattern in the begining of the string.
    - If the pattern is found then return match object. We need to apply gorup() method on match object to get the 
    matching value.
    - If pattern is not found in the begining of the string then it will return None.

search
    - looks for the pattern in the whole string and returns the match object of first occurence.
    - If the pattern is found then return match object. We need to apply gorup() method on match object to get the 
    matching value.
    - If pattern is not found in the begining of the string then it will return None.

findall
    - Looks fro the pattern in whole sting and return a list of all the matching values.
    - If no matched value is foud then it will return None.

finditer
    - Looks fro the pattern in whole sting and return an iterator object of all matching objects in string

split
    - splits the string from the given pattern and returns a list of splitted strings
```
